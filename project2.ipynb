## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed
#save and print the text content of a page with all tags removed
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Jalen_Hurts"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

text_content = soup.get_text()

with open("page_text.txt", "w", encoding="utf-8") as file:
    file.write(text_content)

print(text_content)
```

```python
#download an image with beautifulsoup and save it in this repository
import os
from urllib.parse import urljoin

url = "https://en.wikipedia.org/wiki/Jalen_Hurts"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

image_tag = soup.find("img")

if image_tag:
    image_url = urljoin(url, image_tag["src"])
    img_response = requests.get(image_url)
    image_name = os.path.join(os.getcwd(), "downloaded_image.jpg")

    with open(image_name, "wb") as file:
        file.write(img_response.content)

    print(f"Image downloaded and saved as {image_name}")
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links
url = "https://en.wikipedia.org/wiki/Jalen_Hurts"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
links = soup.find_all("a", href=True)
for link in links[:10]:
    href = link['href']
    full_link = href if href.startswith("http") else f"https://en.wikipedia.org{href}"
    print(full_link[:100])
```

## Part 3: Downloading scripts

```python
scripts=pd.read_csv('pudding_data.csv')
```

```python
scripts
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags

def fetch_script_text(url):
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")
        clean_text = soup.get_text()
        return clean_text[:1000]
    except requests.RequestException as e:
        return f"Error: {e}"

scripts_sample = scripts["link"].dropna().head(5)
script_snippets = {link: fetch_script_text(link) for link in scripts_sample}

for link, text in script_snippets.items():
    print(f"URL: {link}\nText Preview:\n{text}\n{'-'*80}\n")
```

```python
#add a new column to the df with the text downloaded
#save this new dataframe as "pudding_texts.csv"

scripts.loc[scripts["link"].notna(), "script_text"] = scripts["link"].map(script_snippets)
scripts.to_csv("pudding_texts.csv", index=False)
scripts.head(5)
```

#### Browse the documentation at https://developer.themoviedb.org/reference/intro/getting-started. Create an account to authenticate

```python
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 

api_key = 'bc207a29240d15fa9132f025db0e1bb0'
url = 'https://api.themoviedb.org/3/movie/now_playing?api_key=bc207a29240d15fa9132f025db0e1bb0&language=en-US&page=1'

response = requests.get(url)
data = response.json()

movies = []
for movie in data['results']:
    movies.append({
        'title': movie['title'],
        'id': movie['id'],
        'release_date': movie['release_date'],
        'overview': movie['overview'],
        'vote_average': movie['vote_average'],
        'genre_ids': movie['genre_ids']
    })

df = pd.DataFrame(movies)

df.to_csv('now_playing_movies.csv', index=False)
df.head(10)
```

```python
#download the movie posters for 10 of these movies and save them to this repository
import requests
import os
import pandas as pd

API_KEY = "bc207a29240d15fa9132f025db0e1bb0"
SEARCH_URL = "https://api.themoviedb.org/3/search/movie"
IMAGE_BASE_URL = "https://image.tmdb.org/t/p/w500"

os.makedirs("movie_posters", exist_ok=True)

def get_poster_path(movie_title):
    params = {
        "api_key": API_KEY,
        "query": movie_title,
        "language": "en-US"
    }
    response = requests.get(SEARCH_URL, params=params)
    
    if response.status_code == 200:
        results = response.json().get("results", [])
        if results:
            return results[0].get("poster_path")
    return None


df_sample = df.head(10)


poster_paths = {}

for index, row in df_sample.iterrows():
    movie_title = row["title"]
    poster_path = get_poster_path(movie_title)
    
    if poster_path:
        poster_paths[movie_title] = poster_path
    else:
        print(f"No poster found for {movie_title} on TMDb.")

for title, poster_path in poster_paths.items():
    poster_url = IMAGE_BASE_URL + poster_path
    img_response = requests.get(poster_url)

    if img_response.status_code == 200:
        file_path = os.path.join("movie_posters", f"{title.replace(' ', '_')}.jpg")
        with open(file_path, "wb") as file:
            file.write(img_response.content)
        print(f"Downloaded: {file_path}")
    else:
        print(f"Failed to download {title}'s poster.")
```
